```python
import pandas as pd
import numpy as np 
from sklearn import preprocessing

MUSHROOMS=pd.read_csv('mushrooms.csv') 

le=preprocessing.LabelEncoder()
mushrooms=MUSHROOMS.apply(le.fit_transform)

target=mushrooms['class']
data=mushrooms.drop('class',axis=1)
data.head()
```

```python
le.fit(MUSHROOMS['cap-shape'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
le_name_mapping
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(data,target,test_size=0.5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

knn=KNeighborsClassifier(6)
knn.fit(x_train,y_train)
knn.score(x_train,y_train)    
```

```python
y_pred=knn.predict(x_test)

from sklearn.metrics import accuracy_score

score_pred_knn=accuracy_score(y_test,y_pred)
print(score_pred_knn)
```

```python
Y_test=np.array(y_test)
df_knn=pd.DataFrame(Y_test,columns=['ACTUAL'])
df_knn['PREDICTED']=y_pred
df_knn.head(20)
```

```python
from sklearn import naive_bayes

gnb=naive_bayes.GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_train,y_train)
```

```python
y_pred_gnb=gnb.predict(x_test)
score_pred_gnb=accuracy_score(y_test,y_pred_gnb)
print(score_pred_gnb)
```

```python
Y_test_gnb=np.array(y_test)
df_gnb=pd.DataFrame(Y_test_gnb,columns=['ACTUAL'])
df_gnb['PREDICTED']=y_pred_gnb
df_gnb.head(20)
```
